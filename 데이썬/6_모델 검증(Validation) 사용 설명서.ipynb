# 모델 검증(Validation) 사용 설명서

## 1. 모델 검증의 필요성

### 1) 모델 검증이란 ?
- 모델 검증: 모델을 설계한 후 모델을 검증하는 단계


- 모델 검증이 필요한 이유: 모델을 적합하게 잘 설계했는지, 즉 예측을 잘하는지 알기 위해서임

### 2) 표본 내 성능 검증 & 표본 외 성능 검증 (교차검증)
- 기본적으로 훈련 데이터를 기반으로 모델링하고, 테스트 데이터로 해당 모델의 성능을 측정
    - 표본 내 성능 검증: 훈련 데이터의 종속 변수값을 잘 예측하는지 보는 과정
    - 표본 외 성능 검증(교차검증): 테스트 데이터의 종속 변수값을 잘 예측하는지 보는 과정
    
    
- 표본 외 성능 검증(교차검증)은 예측을 잘 하는지 알기 위해서 학습에 사용하지 않는 표본을 활용해 모델이 잘 돌아가는지 확인

## 2.  Train test split

### 1) Train test split이란?
- 교차검증을 하기 위해서 두 종류의 데이터 집합 필요
    - Training data set: 모델 학습을 위한 데이터 집합
    - Test data set: 모델 성능 검증을 위한 데이터 집합
    
    
- Train test split(학습/검증 데이터 분리) 메소드를 사용해 학습용 데이터와 검증용 데이터로 분리

## 3. 교차검증

### 1) 교차검증이란?


- 교차검증: 모델을 평가하는 방법 중 하나로, 데이터 셋을 그림과 같이 나누게 됨


- Valid data를 모델의 과적합을 방지하기 위해 사용


- 보통 Train : Test 데이터 분리 비율은 8:2 ~ 7:3 사이

### 2) K-fold 교차검증이란?


- K-fold 교차검증: Train data set 전체를 K개의 학습세트와 검증세트로 나누어, K만큼의 폴드세트에 K번의 모델 학습과 검증을 수행


- K번의 교차검증 성능을 평균내어 최종 교차검증 성능으로 계산


- 데이터의 수가 적은 경우 사용하기 좋은 방법

### 3) K-fold 교차검증의 장단점
- 교차검증의 장점
    - 모든 데이터셋을 평가에 반영할 수 있어 데이터의 편중을 막을 수 있음
    - 더욱더 일반적인 모델을 생성 가능하여 과적합 방지
    - 반복적인 교차검증을 통해 모델의 정확도를 향상시킬 수 있음
    - 데이터가 부족하여 train과 test로 나누기에 아까운 경우에 활용하기 좋음
    
    
- 교차검증의 단점
    - k번의 모델을 생성해야 하므로, 모델 생성/평가하는 과정에서 시간이 많이 소요

## 4. Metrics

### 1) Metrics란?


- 사이킷런의 Metrics 서브패키지에는 Confusion Matrix라는 훈련된 모델의 성능을 측정하기 위한 Matrix가 있음


- 그림은 Confusion Matrix(이진 분류결과표)
    - TP(True Positive): True를 True로 잘 예측한 것
    - TN(True Negative): False를 False로 잘 예측한 것
    - FP(False Positive): False를 True로 잘 못 예측한 것
    - FN(False Negative): True를 False로 잘 못 예측한 것
    
    
- Confusion Matrix를 활용해 정확도, 정밀도, 재현율 등 측정 가능

### 2)  평가점수
- 사이킷런의 Metrics 서브패키지에는 예측 성능을 평가하기 위한 다양한 함수를 제공


- 그 중 회귀분석에 유용한 함수는 다음과 같음
    - R2_score: 결정계수
    - Mean_squared_error: 평균 제곱 오차
    - Median_absolute_error: 절대 오차 중앙값

#### [ 정확도 (Accuracy) ]

- 정확도: 전체 샘플 중 맞게 예측한 샘플 수의 비율로, 높을수록 좋은 모형

#### [ 정밀도 (Precision) ]

- 정밀도: positive 클래스에 속한다고 예측한 샘플 중 실제로도 positive 클래스에 속하는 샘플 수의 비율로, 높을수록 좋은 모형

#### [ 재현율 (Recall) ]

- 재현율(=민감도): 실제 positive 클래스에 속한 표본 중에 positive 클래스에 속한다고 예측한 표본 수의 비율로, 높을수록 좋은 모형

#### [ 위양성율 (Fall-out) ]

- 위양성율: 실제 negative에 속하지 않는 표본 중에 negative에 속한다고 출력한 표본의 비율로, 1에서 위양성율의 값을 뺀 값을 특이도라고 함
