# 데이터 분석 사용 설명서

## 1. 데이터 분석이란?

데이터 분석(Data Analysis)은 데이터를 정리, 변환, 모델링을 통해 유용한 정보를 발굴하고 결론을 도출하여 유의미한 의사결정을 도와주는 과정이다.

## 2. 데이터 분석의 필요성

### [ 데이터를 가치있게 만드는 과정 ]
- 미래 예측: 데이터 분석을 통해 인사이트를 확인, 미래 상황을 예측


- 데이터 기반의 의사결정: 데이터 분석을 활용하여 객관적인 의사결정 도출


- 추가적인 수익 창출: 데이터 분석을 통해 도출된 결과로 원가를 절감하고 리스크를 통제하며 기업의 수익 극대화


- 효율성의 극대화: 의사결정과 행동에 도움을 줌으로써 전체적인 효율 향상

## 3. 데이터 분석 프로세스

### 1) 문제 정의
- 문제 정의는 데이터 분석을 하는데 있어 주제를 선정하는 과정으로, 어떤 주제를 선정하냐에 따라 결과물에 직접적으로 영향을 주기 때문에 가장 중요한 단계라고 할 수 있다.


- 분석하고자 하는 대상을 정확히 이해하고, 보다 더 객관적이고 구체적으로 정의하는 과정이 필요하다.


- 데이터 수집을 먼저 한 뒤 수집한 데이터에 맞게 문제를 정의하면 편하겠지만 그 데이터에 귀속되어 새로운 접근 방법을 찾아내기 어렵다. 그렇기에 문제 정의 -> 데이터 수집 순으로 단계를 진행하는 것이 좋다.

### 2) 데이터 수집
- 문제 정의 과정을 거친 후 필요한 데이터의 조건을 정의해야 한다.


- 데이터를 어디서 얻을 수 있는지 파악한 후 수집한다.

### 3) 데이터 전처리
#### [ EDA (데이터 탐색) ]
- 데이터를 탐색하고 특성을 파악하여 여러 관계를 찾는 과정


    - 상관 관계
    - 분포 확인
    - 인과 관계
    
#### [ 데이터 전처리 ]
- 오류 사항을 점검하고 데이터의 구조 및 특성을 분석에 용이하게 바꾸는 과정


    - scaling
    - 결측치 & 이상치 처리
    - 범주화

### 4) 데이터 모델링
- 모델링은 원하는 결과(예측, 분류)를 얻기 위해 다양한 관점에서 데이터를 설계하고 테이블 간의 관계를 설정하는 과정이다.


- 다양한 머신러닝 기법(결정 트리 모델, 앙상블 등)을 적용하고 성능을 높이기 위한 파라미터 튜닝 작업을 진행한다.

### 5) 시각화 및 탐색
- 모델링을 통해 나온 결과를 정의한 문제와 연결하여 해결할 수 있는 방법을 인사이트를 도출한다.


- 좀 더 설득력 있는 결과물을 만들어내기 위해 다양한 유형의 시각화를 활용한다.
