## 3. Matplotlib 기초 사용법

### 1) 기본 그래프

```python
# 선그래프
plt.plot([1, 2, 3, 4])
plt.show()
```

```python
# plot() 함수 딕셔너리 데이터 입력
data_dict = {'x': [1, 2, 3, 4], 'y': [1, 3, 7, 13]}
plt.plot('x', 'y', data=data_dict)
plt.show()
```

```python
# 여러 개의 곡선을 하나의 그래프로
import numpy as np
t = np.arange(0, 5, 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### 2) 축 레이블 설정하기

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 13])
plt.xlabel('x_label', labelpad=15)   # labelpad: 축 레이블의 여백
plt.ylabel('y_label', labelpad=15)
plt.show()
```

### 3) 범례 표시하기

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 13], label='Price ($)')
plt.legend(loc='lower right')   # loc: 범례의 위치 설정
plt.show()
```

### 4) 축 범위 지정하기

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 13])
plt.xlim((0, 5))
plt.ylim((0, 16))
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 13])
plt.axis([0, 5, 0, 16])
plt.show()
```

### 5) 선 종류 지정하기
1. 기본 사용: '-' (solid), '--' (dashed), ':' (dotted), '-.' (dashdot)


2. linestyle 파라미터 사용: 'solid', 'dashed', 'dotted', 'dashdot'


3. 튜플 사용: (0, (1, 1)) (), (0, (1, 5)) (dotted), (0, (5, 1)) (dashed), (0, (3, 5, 1, 5)) (dashdot)

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 13], '--')
plt.plot([1, 2, 3, 4], [2, 4, 8, 14], linestyle='dashdot')
plt.plot([1, 2, 3, 4], [3, 5, 9, 15], linestyle=(0, (1, 5)))
```

### 6) 마커 지정하기

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 13], 'ro')
plt.plot([1, 2, 3, 4], [2, 4, 8, 14], 'bo--')
plt.plot([1, 2, 3, 4], [3, 5, 9, 15], marker='v')
```

### 7) 색상 지정하기
1. plot() 함수의 포맷 문자열 사용 ex) 'r', 'b', 'g'


2. color 키워드 인자 사용 ex) color='violet', color='red'


3. Hex code 사용 ex) color='#5ecbc8'

```python
plt.plot([1, 2, 3, 4], [1, 3, 7, 13], 'r')
plt.plot([1, 2, 3, 4], [2, 4, 8, 14], color='green')
plt.plot([1, 2, 3, 4], [3, 5, 9, 15], color='#5ecbc8')
```
